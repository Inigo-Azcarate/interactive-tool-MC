```python
trips_ez = pd.read_csv('workers_eskuzaitzeta_2k.csv')
```

## CREACIÃ“N DE ARCHIVOS HEXAGON Y ARCS

```python
# Hexagon

trips_ez['COORDINATES'] = trips_ez.apply(lambda row: [row['O_long'], row['O_lat']], axis=1)

# Luego, seleccionamos solo la columna 'COORDINATES' y la convertimos en una lista de diccionarios
coordinates_list = trips_ez['COORDINATES'].apply(lambda x: {'COORDINATES': x}).tolist()

# Finalmente, guardamos esta lista en un archivo JSON
with open('heatmap_layer.json', 'w') as f:
    json.dump(coordinates_list, f)

# Mostramos el resultado para verificar
print(coordinates_list[:5])
```

```python
# Arcs

def determine_inbound(mode):
    if mode == "Car":
        return 1
    elif mode == "PT":
        return 2
    elif mode == "Andando":
        return 3
    else:
        return 0

json_list = [
    {
        "inbound": determine_inbound(row['Modo']),
        "from": {
            "coordinates": [row['O_long'], row['O_lat']]
        },
        "to": {
            "coordinates": [row['D_long'], row['D_lat']]
        }
    } for index, row in trips_ez.iterrows()
]

# Save the JSON structure to a file
with open('arc_layer.json', 'w') as outfile:
    json.dump(json_list, outfile, indent=2)
```
