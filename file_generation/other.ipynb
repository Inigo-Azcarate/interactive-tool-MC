```python
results_total_sum = []
results_mode_counts = []
results_total_distances = []

for file in datasets:
    trips_ez = pd.read_csv(file)

    trips_ez['distance'].replace(0.000, 3000, inplace=True)

    factor = {'Car': 0.000162, 'PT': 0.00008, 'Andando': 0} # T/km or #kg/m. It's the same.
    trips_ez['multiplied_distance'] = trips_ez.apply(lambda row: row['distance'] * factor.get(row['prediction'], 1), axis=1)
    total_sum = trips_ez['multiplied_distance'].sum()
    total_sum = math.sqrt(total_sum/3.141592)

    mode_counts = trips_ez['prediction'].value_counts()
    array_mode_counts = [mode_counts.get('Car', 0), mode_counts.get('PT', 0), mode_counts.get('Andando', 0)]

    total_distances = trips_ez.groupby('prediction')['distance'].sum()
    array_total_distances = [total_distances.get('Car', 0)/1000, total_distances.get('PT', 0)/1000, total_distances.get('Andando', 0)/1000]

    results_total_sum.append(total_sum)
    results_mode_counts.append(array_mode_counts)
    results_total_distances.append(array_total_distances)
```
